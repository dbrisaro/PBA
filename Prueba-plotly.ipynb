```python
import pandas as pd

path = '/home/daniu/Documentos/PBA/mapas/'

filename = 'Base de datos mapa.xlsx'

df_uni = pd.read_excel(path + filename, sheet_name=0, index_col=0)

latitud_uni = df_uni['LATITUD'].values
longitud_uni = df_uni['LONGITUD'].values
nombres_uni = df_uni['ID'].values

df_inst = pd.read_excel(path + filename, sheet_name=1, index_col=0)

latitud_inst = df_inst['LATITUD'].values
longitud_inst = df_inst['LONGITUD'].values
nombres_inst = df_inst['ID'].values

df_obs = pd.read_excel(path + filename, sheet_name=2, index_col=0)

latitud_obs = df_obs['LATITUD'].values
longitud_obs = df_obs['LONGITUD'].values
nombres_obs = df_obs['ID'].values
```

```python
import plotly.graph_objects as go
import plotly.offline as pyo

mapbox_access_token = open(".mapbox_token").read()

fig = go.Figure(go.Scattermapbox(
        lat=latitud_uni,
        lon=longitud_uni,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=15,
            color='rgb(177, 0, 0)',
            opacity=0.7
        ),
        text=nombres_uni,
        name='Universidades',
        hovertemplate = "%{text}"
    ))

fig.add_trace(go.Scattermapbox(
        lat=latitud_inst,
        lon=longitud_inst,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=15,
            color='rgb(0, 0, 177)',
            opacity=0.7
        ),
        text=nombres_inst,
        name='Instituciones',
        hovertemplate = "%{text}"
    ))

fig.add_trace(go.Scattermapbox(
        lat=latitud_obs,
        lon=longitud_obs,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=15,
            color='rgb(0, 177, 172)',
            opacity=0.7
        ),
        text=nombres_obs,
        name='Observatorios',
        hovertemplate = "%{text}"
    ))


fig.update_layout(
    autosize=True,
    width=1400,
    height=800,
    hovermode='closest',
    mapbox=dict(
        accesstoken=mapbox_access_token,
        bearing=0,
        center=dict(
            lat=-35.50,
            lon=-61.50
        ),
        pitch=0,
        zoom=6,
        style='light' # 'outdoors'
    ),
)

fig.show()
fig.write_html(path + "Iniciativas-genero.html")
```
